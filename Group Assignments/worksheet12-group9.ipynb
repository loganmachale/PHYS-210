## A. The trainee
You are ready to go on more advanced missions for the USS Enterprise, however you have to train someone in python to continue your work. They have written a function that takes two numbers, x and y and returns $x*\sin y$. It also prints a message before returning. The trainee has a draft, but needs help to make it more "error-proof." They have commented the areas needing help. Rewrite the code using try-except-else statements to ensure the function will not crash with invalid input.

```python
import numpy as np

def x_siny():

    x = input('enter your x value: ') # How to make sure it's a numer?
    y = input('enter your y value: ') # Here too?

    try:
        float(x)
        float(y)
    except:
        print("error converting to floats")
    else:
        x = float(x)
        y = float(y)
        ans = (x*np.sin(y))
        print('x(sin(y)) = ',ans)
        return ans
    

    
x_siny()
```

Create a greyscale version of the image: this is a new 3d array in which the red, green and blue channels for each pixel all have identical values, equal to the average of the red, green and blue channels for the pixels in the original. You should not need to use any loops, just a few lines with some array indexing/broadcast magic.

**Note:** imshow expects that for images represented as integers, each integer should have values 0-255. Images where the pixel intensities are integers, the values should be in the range of 0-1

```python
#  Read the image
from imageio import imread
import matplotlib.pyplot as plt

a = imread('worksheet12-vancouver2.jpeg')
print(a.shape)
plt.imshow(a)
plt.show
```

```python
b = np.copy(a)
b = a*np.array([0,1,0])
print(np.shape(b))
plt.imshow(b)
plt.show
```

```python
c = np.copy(a)
m = np.add(c[:,:,0],c[:,:,1],c[:,:,2])
m = np.dstack((m,m,m))

plt.imshow(m)
plt.show
```
